### Use the marco geoestad√≠stico 2010 (https://www.inegi.org.mx/app/biblioteca/ficha.html?upc=889463807469)     and 2020 (https://www.inegi.org.mx/app/biblioteca/ficha.html?upc=702825292812) to obtain the ‚Äú√Åreas Geoestad√≠sticas B√°sicas‚Äù(AGEBs) from M√©rida, Yucat√°n. AGEBs evolve on time. Use sets to find:

- `geopandas` is imported with the alias `gpd`.
- Two variables, `file_mex_2010` and `file_yuc_2020`, are defined, each holding the path to a specific Shapefile.

```python
import geopandas as gpd

file_mex_2010 = '2010_MEX\AGEB_urb_2010_5.shp'
file_yuc_2020 = '2020_YUC\conjunto_de_datos\\31a.shp'
```

This code segment uses the geopandas library to read a Shapefile and load the geospatial data into a GeoDataFrame. 

```python
mex_2010 = gpd.read_file(file_mex_2010)
mex_2010
```

This code snippet is manipulating a `GeoDataFrame` named `mex_2010`, which contains geospatial data previously loaded from a Shapefile. Specifically, it's extracting and creating a new column based on existing data:

```python
mex_2010["CVE_ENT"] = mex_2010['CVEGEO'].str.slice(0,2)
mex_2010
```

This line of code continues the manipulation of the `mex_2010` GeoDataFrame, which contains geospatial data.

```python
mex_2010["CVE_MUN"] = mex_2010["CVEGEO"].str.slice(2,5)
mex_2010
```

This code snippet filters the `mex_2010` GeoDataFrame to select only the rows where the `"CVE_ENT"` column equals `'31'`, which represents the entities within the state of Yucat√°n, given that '31' is the code for Yucat√°n in Mexico's geographical coding system. The filtered data is then assigned to a new GeoDataFrame named `yuc_2010_data`.

```python
yuc_2010_data = mex_2010[mex_2010['CVE_ENT'] == '31']
yuc_2010_data
```

This line of code further filters the `yuc_2010_data` GeoDataFrame, narrowing down the selection to only include the rows where the `"CVE_MUN"` column equals `'050'`. This specific operation isolates data for a particular municipality within the state of Yucat√°n, with the municipality code `'050'` referring to M√©rida, the capital and largest city of Yucat√°n. The filtered data is then stored in a new GeoDataFrame named `merida_2010`.

```python
merida_2010 = yuc_2010_data[yuc_2010_data["CVE_MUN"] == "050"]
merida_2010
```

This snippet of code reads a Shapefile related to Yucat√°n for the year 2020 and stores its data in a `GeoDataFrame` named `yuc_2020`:

```python
yuc_2020 = gpd.read_file(file_yuc_2020)
yuc_2020
```

This line of code filters the `yuc_2020` GeoDataFrame to select only the rows where the `"CVE_MUN"` column equals `'050'`, which likely represents the municipality code for M√©rida in the state of Yucat√°n for the year 2020. The filtered data is then assigned to a new GeoDataFrame named `merida_2020`.

```python
merida_2020 = yuc_2020[yuc_2020["CVE_MUN"] == "050"]
merida_2020
```

This code segment performs a set operation on two GeoDataFrames (`merida_2010` and `merida_2020`), extracting the unique 'CVEGEO' values for M√©rida from two different years (2010 and 2020) and printing these sets for comparison:

1. **Create Sets of Unique Identifiers**:
   - `merida_2010_set = set(merida_2010['CVEGEO'])`: Converts the 'CVEGEO' column values from the `merida_2010` GeoDataFrame into a set, effectively removing any duplicates and storing unique geographic codes for M√©rida in 2010.
   - `merida_2020_set = set(merida_2020['CVEGEO'])`: Does the same for the `merida_2020` GeoDataFrame, storing unique geographic codes for M√©rida in 2020.

2. **Print Sets**:
   - The `print` statements then display these sets, labeled as `merida_2010_set` and `merida_2020_set`, respectively. This provides a clear textual representation of the unique geographic identifiers used in both years, allowing for a straightforward comparison of what has remained the same, what has changed, or what has been added or removed between 2010 and 2020.

```python
merida_2010_set = set(merida_2010['CVEGEO'])
merida_2020_set = set(merida_2020['CVEGEO'])


print("merida_2010_set:", merida_2010_set)
print("merida_2020_set:", merida_2020_set)
```

This snippet of code uses set operations to analyze the differences and similarities in geographic codes ('CVEGEO') for M√©rida between the years 2010 and 2020, effectively identifying which areas remained unchanged, which are new, and which disappeared over the decade.

```python
remains_2010_2020 = merida_2010_set & merida_2010_set
new_in_2020 = merida_2020_set - merida_2010_set
disappears_2010_2020 = merida_2010_set - merida_2020_set
```

This code snippet filters the `merida_2020` GeoDataFrame to create a new GeoDataFrame, `remains_2010_2020_gdf`, which contains only the areas that remained unchanged between 2010 and 2020 in M√©rida, based on their 'CVEGEO' geographic codes.

```python
remains_2010_2020_gdf = merida_2020[merida_2020["CVEGEO"].isin(remains_2010_2020)]
remains_2010_2020_gdf
```

This code snippet is creating a new GeoDataFrame named `new_in_2020_gdf` by filtering the `merida_2020` GeoDataFrame. It includes only those areas that are new in M√©rida for the year 2020, based on their 'CVEGEO' geographic codes, which did not appear in the 2010 dataset.

```python
new_in_2020_gdf = merida_2020[merida_2020["CVEGEO"].isin(new_in_2020)]
new_in_2020_gdf
```

This line of code filters the `merida_2010` GeoDataFrame to create a new GeoDataFrame, `disappears_2010_2020_gdf`, which includes only those areas that disappeared from M√©rida between the years 2010 and 2020. This is based on their 'CVEGEO' geographic codes, which were present in the 2010 dataset but not in the 2020 dataset.

```python
disappears_2010_2020_gdf = merida_2010[merida_2010["CVEGEO"].isin(disappears_2010_2020)]
disappears_2010_2020_gdf
```

This code snippet uses `matplotlib` and `geopandas` to create a visual map displaying the changes in the AGEBs (√Åreas Geoestad√≠sticas B√°sicas) of M√©rida from 2010 to 2020. It categorizes the AGEBs into three groups based on whether they remained unchanged, appeared new in 2020, or disappeared from 2010 to 2020, and plots each category with distinct colors for clear differentiation.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# Plot your data with updated colors
remains_2010_2020_gdf.plot(ax=ax, color='blue', edgecolor='white', linewidth=0.5, alpha=0.8, label='Remains 2010-2020')  # Remains in light blue
new_in_2020_gdf.plot(ax=ax, color='purple', edgecolor='white', linewidth=0.5, alpha=0.8, label='New in 2020')  # New in purple
disappears_2010_2020_gdf.plot(ax=ax, color='magenta', edgecolor='white', linewidth=0.5, alpha=0.8, label='Disappeared 2010-2020')  # Disappeared in pink

# Create custom legend with the updated colors
legend_labels = [
    Patch(facecolor='lightblue', edgecolor='white', label='Remains 2010-2020'),
    Patch(facecolor='purple', edgecolor='white', label='New in 2020'),
    Patch(facecolor='pink', edgecolor='white', label='Disappeared 2010-2020')
]
ax.legend(handles=legend_labels)

# Customizations
ax.set_title('AGEB changes from 2010 to 2020')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.grid(True)

plt.show()
```
